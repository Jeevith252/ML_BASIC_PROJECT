## **Ecoding - Converting Categorical into Numerical**


### **Two methods - LabelEncoder and OneHotEncoder**





**1)Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
le = LabelEncoder()
```

```python
df = pd.read_csv("/content/customer_churn_dataset-training-master.csv")
```

```python
df.head()
```

```python
df["Gender"] = le.fit_transform(df["Gender"])
```

```python
df["Gender"]
```

**2)One Hot Encoder**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
S_T = ohe.fit_transform(df[["Subscription Type"]])
```
